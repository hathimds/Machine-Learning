### Data Preprocessing

#### Convert Attrition from ('Yes', 'No') to (1,0) 

```python
df['Attrition'] = pd.get_dummies(df['Attrition'], dtype=int, drop_first=True)
```

```python
df.head()
```

### Convert the rest of the categorical values into numeric using dummy variables and store the results in a new dataframe called 'newdf'

```python
newdf = pd.get_dummies(df, dtype=int, drop_first=True)
```

```python
newdf.head()
```

#### Check the shape of our new dataset

```python
newdf.shape
```

#### Print unique values in our new dataframe

```python
for feature in newdf:
    print(f"Unique values in {feature} are: {newdf[feature].unique()}")
```

#### Split the columns into input and target variables (X and y)

```python
X = newdf.drop(['Attrition'], axis=1)
y = newdf['Attrition']
```

#### Carry out Feature scaling using StandardScaler

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
xcolumns = X.columns
```

```python
X = scaler.fit_transform(X)
X = pd.DataFrame(X, columns=xcolumns)
```

```python
X.head()
```

### Split the dataset into training and testing set

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=100)
```

## Machine Learning Models

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.model_selection import cross_val_score
```

```python
lr_model = LogisticRegression()
```

```python
lr_model.fit(X_train, y_train)
```

```python
y_pred = lr_model.predict(X_test)
```

```python
lr_accuracy_score = metrics.accuracy_score(y_test, y_pred)
lr_confusion_matrix = metrics.confusion_matrix(y_test, y_pred)
lr_classification_report = metrics.classification_report(y_test, y_pred)
lr_cross_val_score = cross_val_score(lr_model, X, y)

print(f"Accuracy score of our prediction is: {accuracy_score}")
print('Confusion matrix is:')
print(confusion_matrix)
print('Classification Report is:')
print(classification_report)
print(f'Cross val score: {lr_cross_val_score.mean():.5f}')
```
